```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Boltzmann's constant, J/K
kB = 1.381e-23

# The Lennard-Jones parameters:
#A = 1.024e-23   # J.nm^6
#B = 1.582e-26   # J.nm^12

#He
#eps = 10.22 #K
#sigma = 0.256 #nm
#Ne
#eps = 35.6 #K
#sigma = 0.275 #nm
#Kr
eps = 164 #K
sigma = 0.383 #nm
A = 4*eps*sigma**6
B = 4*eps*sigma**12

# Adjust the units of A and B - they have more manageable values
# in K.nm^6 and K.nm^12
#A, B = A / kB, B / kB

# Interatomic distance, in nm
r = np.linspace(0.1, 1, 1000)
# Interatomic potential
U = B/r**12 - A/r**6
# Interatomic force
F = -12*B/r**13 + 6*A/r**7

line1 = plt.plot(r, U, 'k', lw=2, label=r'U(r)')
plt.xlim(0.2, 0.8)
plt.ylim(-200, 200)
plt.ylabel('Potential U(r)')
plt.xlabel('Distance (nm)')
plt.twinx()
line2 = plt.plot(r, F, 'k', ls=':', lw=2, label=r'F(r)')
plt.xlim(0.2, 0.8)
plt.ylim(-1000, 1500)
plt.ylabel('Force F(r)')

lines = line1 + line2
labels = []
for line in lines:
    labels.append(line.get_label())
plt.legend(lines, labels)

#plt.show()
plt.savefig('P1_Ne.png')
```
